```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.color_palette("bright");
```

```python
X = np.load("../data/processed_extrapolation.npy", allow_pickle=True)
E = X[()]['data'][:,1:]
h_omega = np.array(X[()]['data'][:,0])
N_Max = X[()]['Nmax'].reshape([-1])
n_points = h_omega.shape[0]

plt.plot(h_omega, E)
plt.xlabel(r'$\bar{h}\omega$')
plt.ylabel(r'$E$')
plt.legend(N_Max)
```

```python
plt.plot(N_Max, E[5,:])
plt.ylabel(r'$E$')
plt.xlabel(r'$N_{max}$')
plt.title(r'$\bar{h}\omega = %g$' % h_omega[5])
```
